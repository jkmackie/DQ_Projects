```python
fcc['MoneyForLearning'].describe()
```

```python
#Replace 0 months programming with 1
fcc['MonthsProgramming'].replace(to_replace=0,value=1,inplace=True)
```

```python
fcc['MoneyPerMo']=fcc['MoneyForLearning']/fcc['MonthsProgramming']
fcc.shape
```

```python
#Filter certain nulls from fcc dataframe
#Remove nulls from MoneyPerMo and CountryLive
fcc=fcc[fcc['MoneyPerMo'].notnull()]
fcc=fcc[fcc['CountryLive'].notnull()]
```

```python
#Remove fcc MoneyPerMo outliers
fcc=fcc[fcc['MoneyPerMo']<20000]
```

```python
#Group dataframe by CountryLive Averages
fcc_mean=fcc.groupby(by=['CountryLive']).mean()
fcc_median=fcc.groupby(by=['CountryLive']).median()
```

```python
#Filter pivot using [PickColumn][[RowName1,Rowname2,etc]]
fcc_mean['MoneyPerMo'][['Canada','India','United Kingdom','United States of America']]
```

```python
fcc_median['MoneyPerMo'][['Canada','India','United Kingdom','United States of America']]
```

```python
import seaborn as sns
```

```python
# Isolate again the countries of interest
only_4 = fcc[fcc['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada')]

# Box plots to visualize distributions
sns.boxplot(y = 'MoneyPerMo', x = 'CountryLive',
            data = only_4)
plt.title('Money Spent Per Month Per Country\n(Distributions)',
         fontsize = 14)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

```python
# Inspect the extreme outliers for India
#Use boolean filtering
india_outliers = only_4[
    (only_4['CountryLive'] == 'India') & 
    (only_4['MoneyPerMo'] >= 2500)]
india_outliers
```

```python
#Show indices of India outlier
india_outliers.index
```

```python
#Remove India outliers by keying off the outlier indices
print(only_4.shape)
only_4=only_4.drop(index=india_outliers.index,errors='ignore')
print(only_4.shape)
```

```python
us_outliers=only_4[
    (only_4['CountryLive']=='United States of America') &
    (only_4['MoneyPerMo']>=6000) &
    (only_4['AttendedBootcamp']==0)]

us_outliers
```

```python
#Remove USA outliers by keying off the outlier indices
print(only_4.shape)
only_4=only_4.drop(index=us_outliers.index,errors='ignore')
print(only_4.shape)
```

```python
# Box plots to visualize distributions
print("only_4 shape: ",only_4.shape)
sns.boxplot(y = 'MoneyPerMo', x = 'CountryLive',
            data = only_4)
plt.title('Money Spent Per Month Per Country\n(Distributions)',
         fontsize = 14)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

```python
# Examine the extreme outliers for Canada
canada_outliers = only_4[
    (only_4['CountryLive'] == 'Canada') & 
    (only_4['MoneyPerMo'] > 4500)]

canada_outliers
```

```python
#drop Canada outliers
only_4 = only_4.drop(canada_outliers.index)
```

```python
print('After outlier removal, coutries with most MoneyPerMo')
print('are USA and Canada:')
only_4_mean=only_4.groupby(by=['CountryLive']).mean()
only_4_mean['MoneyPerMo'][['Canada','India','United Kingdom','United States of America']]
```
